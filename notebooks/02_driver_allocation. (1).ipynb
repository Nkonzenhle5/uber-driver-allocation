```python
import pandas as pd

# ==============================
# LOAD DATA
# ==============================
df = pd.read_csv("results/demand_predictions_advanced.csv")

# ==============================
# PARAMETERS
# ==============================
drivers = 500
hours_per_driver = 8
total_driver_hours = drivers * hours_per_driver

# ==============================
# ALLOCATION
# ==============================
total_demand = df['predicted_rides'].sum()

df['allocated_driver_hours'] = (df['predicted_rides'] / total_demand) * total_driver_hours
df['drivers_allocated'] = (df['allocated_driver_hours'] / hours_per_driver).round().astype(int)

# ==============================
# SAVE OUTPUT
# ==============================
df.to_csv("results/driver_allocation_output.csv", index=False)

# ==============================
# PRINT OUTPUT
# ==============================
print("\n===== DRIVER ALLOCATION =====")
for i, row in df.iterrows():
    print(f"Hour {int(row['hour'])}: Demand = {int(row['predicted_rides'])}, Drivers = {row['drivers_allocated']}")
```
